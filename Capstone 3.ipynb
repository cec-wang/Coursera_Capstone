### Part 2: data processing for collaborative filtering

```python
# one hot encoding
M9R_onehot = pd.get_dummies(M9R_venues[['Venue Category']], prefix="", prefix_sep="")

#Group rows by postal codes and take the mean of the frequency of occurrence of each category
M9R_grouped=M9R_onehot.mean().reset_index()
M9R_grouped.iloc[0,1]=0.2

M9R_grouped.head()
```

```python
# Add a new list to assign each venue an ID number
id_list_short=np.arange(204).tolist()
```

```python
 #create a new dataframe
Toronto_venues3 = pd.DataFrame(columns=['Postal Code','Venue Categories','Mean','Category ID'])

PostalCode_list=[]
Categories_list=[]
Mean_list=[]
id_list=[]

for i in np.arange(39):
    Categories_list=Categories_list+Categories_list_short
    Mean_list=Mean_list+Toronto_grouped2.iloc[2:,i].tolist()
    id_list=id_list+id_list_short
    for m in np.arange(204):
        PostalCode_list.append(Toronto_grouped2.iloc[1,i])
```

```python
 Toronto_venues3['Postal Code']=PostalCode_list
Toronto_venues3['Venue Categories']=Categories_list
Toronto_venues3['Mean']=Mean_list
Toronto_venues3['Category ID']=id_list
Toronto_venues3.shape
```

```python
# Add Category ID to M9R
inputId = VenueID[VenueID['Venue Categories'].isin(M9R_grouped['Venue Categories'].tolist())]
M9R_grouped = pd.merge(inputId, M9R_grouped)
M9R_grouped
```

```python
#Get the subset of postal codes that have the same venue category
PCSubset = Toronto_venues3[Toronto_venues3['Category ID'].isin(M9R_grouped['Category ID'].tolist())]
PCSubset.head()
```

```python
PCSubsetGroup = PCSubset.groupby(['Postal Code'])

#Sorting it so postal codes with venue most in common with the input will have priority
PCSubsetGroup = sorted(PCSubsetGroup,  key=lambda x: len(x[1]), reverse=True)
```

### Part 3: calculate Pearson correlation coeff.

```python
from math import sqrt
```

Calculate the Pearson correlation between input user and subset group, and store it in a dictionary.

```python
#Store the Pearson Correlation in a dictionary, where the key is the user Id and the value is the coefficient
pearsonCorrelationDict = {}

#For every user group in our subset
for name, group in PCSubsetGroup:
    #Let's start by sorting the input and current user group so the values aren't mixed up later on
    group = group.sort_values(by='Category ID')
    M9R_grouped = M9R_grouped.sort_values(by='Category ID')
    #Get the N for the formula
    nRatings = len(group)
    #Get the review scores for the movies that they both have in common
    temp_df = M9R_grouped[M9R_grouped['Category ID'].isin(group['Category ID'].tolist())]
    #And then store them in a temporary buffer variable in a list format to facilitate future calculations
    tempRatingList = temp_df['Mean'].tolist()
    #Let's also put the current user group reviews in a list format
    tempGroupList = group['Mean'].tolist()
    #Now let's calculate the pearson correlation between two users, so called, x and y
    Sxx = sum([i**2 for i in tempRatingList]) - pow(sum(tempRatingList),2)/float(nRatings)
    Syy = sum([i**2 for i in tempGroupList]) - pow(sum(tempGroupList),2)/float(nRatings)
    Sxy = sum( i*j for i, j in zip(tempRatingList, tempGroupList)) - sum(tempRatingList)*sum(tempGroupList)/float(nRatings)
    
    #If the denominator is different than zero, then divide, else, 0 correlation.
    if Sxx != 0 and Syy != 0:
        pearsonCorrelationDict[name] = Sxy/sqrt(Sxx*Syy)
    else:
        pearsonCorrelationDict[name] = 0
```

```python
pearsonDF = pd.DataFrame.from_dict(pearsonCorrelationDict, orient='index')
pearsonDF.columns = ['similarityIndex']
pearsonDF['Postal Code'] = pearsonDF.index
pearsonDF.index = range(len(pearsonDF))
pearsonDF.head()
```

```python
topUsers=pearsonDF.sort_values(by='similarityIndex', ascending=False)
topUsers.head()
```

Unfortunately, there is **very weak** pearson correlation between postal code M9R and the inner Toronto area.
