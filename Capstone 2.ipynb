### 3. perform density-based clustering by general category.

```python
# one hot encoding
Toronto_onehot2 = pd.get_dummies(Toronto_venues[['Category']], prefix="", prefix_sep="")

# add postal code column back to dataframe
Toronto_onehot2['Postal Code'] = Toronto_venues['Postal Code'] 

# move postal code column to the first column
fixed_columns = [Toronto_onehot2.columns[-1]] + list(Toronto_onehot2.columns[:-1])
Toronto_onehot2 = Toronto_onehot2[fixed_columns]
```

```python
Toronto_onehot2.head()
```

```python
Toronto_cat_grouped=Toronto_onehot2.groupby('Postal Code').mean().reset_index()
Toronto_cat_grouped.head(10)
```

Cluster the postal codes by categories. I will be trying out density-based clustering.  

```python
from sklearn.cluster import DBSCAN 
from sklearn.preprocessing import StandardScaler 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
Toronto_grouped_DBSCN=Toronto_cat_grouped.drop('Postal Code',1)
```

```python
epsilon = 0.25
minimumSamples = 2
db = DBSCAN(eps=epsilon, min_samples=minimumSamples).fit(Toronto_grouped_DBSCN)
labels = db.labels_
labels
```

Add labels back to the table. 

```python
Toronto_cat_grouped.insert(0, 'Cluster Labels', labels)
```

Print postal codes with top 5 categories.

### 4. Examine clustering results.

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_categories = 5

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Postal Code']
for ind in np.arange(num_top_categories):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

```

```python
# create a new dataframe
neighborhoods_categories_sorted = pd.DataFrame(columns=columns)
neighborhoods_categories_sorted['Postal Code'] = Toronto_cat_grouped['Postal Code']

for ind in np.arange(Toronto_cat_grouped.shape[0]):
    neighborhoods_categories_sorted.iloc[ind, 1:] = return_most_common_venues(Toronto_cat_grouped.iloc[ind,1:], 5)

neighborhoods_categories_sorted.head()
```

```python
neighborhoods_categories_sorted.insert(0, 'Cluster Labels', labels)
```

### 5. Plot the clusters

```python
# !conda install -c conda-forge folium=0.5.0 --yes 
import folium # map rendering library

#import all necessary libraries
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
#merge the labels with Longitude and latitude
Toronto_merged = pd.merge(Toronto2, Toronto_cat_grouped, left_on='Postal Code', right_on= 'Postal Code')
```

```python
Toronto_merged.drop('Unnamed: 0', axis=1)
```

```python
# create map
map_clusters = folium.Map(location=[43.666659, -79.381472], zoom_start=11)

# set color scheme for the clusters
x = np.arange(6)
ys = [i + x + (i*x)**2 for i in range(6)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(Toronto_merged['Latitude'], Toronto_merged['Longitude'], Toronto_merged['Postal Code'], Toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
